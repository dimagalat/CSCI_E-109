```python
# Change this list to include more countries. The entries must be
# in the same format used by the World Bank: http://data.worldbank.org/node/18

#COUNTRIES = ["EAS", "chn"] # List of countries and/or regions
WORLD = ["EAS", "ECS", "LCN", "MEA"]
```

```python
# get_page(string, int)
#
# This function fetches a page from the World Bank API.
#
# Input:
#   endpoint string: The REST endpoint for the query.
#   page int: The page number to fetch. Defaults to the first one.
#
# Output:
#   tuple: A tuple with the metadata and the page data.
#
# Example:
#   _get_page("countries/br", 2)
#   ({u'page': 2, u'pages': 1, u'per_page': u'100', u'total': 1}, [])
#
def _get_page(endpoint, page=1):
    url = "http://api.worldbank.org/{}"
    payload = {"format": "json", "per_page": 100, "page": page} # parameters for GET
    response = requests.get(url.format(endpoint), params=payload).json()
    metadata, data = response
    return metadata, data
```

```python
# get_indicator(string, int)
#
# This function fetches all (multi-page) indicators.
#
# Input:
#   country string: The countries and/or regions of interest.
#   indicator string: The indicator of interest.
#
# Output:
#   list: A list of dict with the data.
#
# Example:
#   _get_indicator("EAS;chn", "IP.PAT.RESD")
#   [...]
#
def _get_indicator(country, indicator):
    endpoint = "countries/{:s}/indicators/{:s}".format(country, indicator)
    # Get the first page.
    metadata, data = _get_page(endpoint)
    indicator_data = data
    next_page = metadata["page"] + 1
    num_pages = metadata["pages"]
    total_values = metadata["total"]
    # Loop over every page.
    while next_page <= num_pages:
        metadata, data = _get_page(endpoint, next_page)
        indicator_data += data
        next_page = metadata["page"] + 1
    # Make sure we got all reults.
    assert len(indicator_data) == total_values
    return indicator_data
```

```python
# create_dataframe(countries)
#
# This function fetches all (multi-page) indicators.
#
# Input:
#   country list strings: The countries and/or regions of interest.
#
# Output:
#   pandas.DataFrame: A dataframe with patents and growth columns and year indexes.
#
# Example:
#   create_dataframe(["EAS", "chn"])
#   [...]
#
def create_dataframe(countries):
    # Pull patent and growth data from website.
    patents = _get_indicator(";".join(countries), "IP.PAT.RESD")
    growths = _get_indicator(";".join(countries), "NY.GDP.MKTP.KD.ZG")
    # fdi = _get_indicator(";".join(countries), "BX.KLT.DINV.CD.WD") # how to get new pages
    # Create generator from dicts. Make sure year and country match.
    rows = ((patent["date"],
             patent["country"]["value"],
             patent["value"],
             growth["value"])
            for patent, growth in zip(patents, growths)
            if patent["date"] == growth["date"]
            and patent["country"]["value"] == growth["country"]["value"])
    # Create dataframe from generator and remove rows with None.
    columns=["year", "country", "patents", "growth"]
    df = pd.DataFrame(rows, columns=columns)
    df.dropna(inplace=True)
    df.reset_index(drop=True, inplace=True)
    # Covert strings to numeric types.
    df.year = df.year.astype(np.int)
    df.patents = df.patents.astype(np.int)
    df.growth = df.growth.astype(np.float)
    return df
    
```

### 2.1 Revising and Enforcing Intellectual Property: ###
For this session we are focusing on policy recommendation 1, namely intellectual property.

There are three different types of intellectual property each with different functions and uses:
1. Patents
2. Copyright (including Integrated Circuit)
3. Trademark

Patents. They are used mainly by the hardware industry. A patent may be seen as a more 'potent' version than a copyright, since it disallows second mover from implementing the same technological process even if they have not seen the patented process or product.

Copyright. Copyright holders do not have the claim to sue users of the particular technological process or product they copyrighted if the  user have not yet seen the copyrighted material.

Trademark. They are used mainly by the website and e-commerce industry. User may prohibit others from using the brand name.

So what can Indonesia do to improve its patents performance?

Technical Correctness:
Indonesia currently has Law Number 14 Year 2001 in regardnig patents. The law is based on best practices from the World Trade Organization and World Intellectual Property Organization. Philosophically it's based on the first-to-file rather than first-to-invent system (unlike the US prior to 2013). There are different debates as to whether which patent philosophy is better. On one aspect, a first-to-file system requires minimal administrative capability by the Patent Office, since the first person to file to the patent office is eligible for the patent. However, such a system has the downside of bias towards the large multinational companies (and hence discourage the innovation of smaller firms which do not have an inhouse legal division).
