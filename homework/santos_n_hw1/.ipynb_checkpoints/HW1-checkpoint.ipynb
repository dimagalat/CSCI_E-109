#### Problem 1(d)

How would you graphically display the relationship between total wins and total salaries for a given year? What kind of plot would be best? Choose a plot to show this relationship and specifically annotate the Oakland baseball team on the on the plot. Show this plot across multiple years. In which years can you detect a competitive advantage from the Oakland baseball team of using data science? When did this end? 

**Hints**: Use a `for` loop to consider multiple years. Use the `teamID` (three letter representation of the team name) to save space on the plot.  

```python
# Helper function to fit the model.
def fit_model(group):
    y = group.W
    X = group.salary
    # The model needs an intercept, so add column of 1s.
    sm.add_constant(X)
    # Fit the model.
    return sm.OLS(y, X).fit()
```

```python
# Create text plot.
def make_plot(year, group):
    source = ColumnDataSource(group)
    source.add(fit_model(group).predict(), name='predicted')
    source.add(source.data['W'] - source.data['predicted'], name='residuals')
    # Get the position of OAK on the data source lists.
    oak_pos = source.data['teamID'].index('OAK')
    
    # Default properties for text plots.
    text_props = dict(angle=0, plot_width=600, plot_height=600, tools='',
                  text_align="center", text_font_size="8pt")
    
    fit_plot = text(x=dict(field='salary', units='data'), y=dict(field='W', units='data'),
                    text=dict(field='teamID', units='data'), source=source, text_color='green',
                    title='Fit - {}'.format(year), **text_props)
    
    fit_plot.line(x='salary', y='predicted', source=source, line_width=2)
    
    xaxis().axis_label='salaries (millions of Dollars)'
    yaxis().axis_label='number of wins'
    grid().grid_line_color='white'
    
    resid_plot = text(x=dict(field='salary', units='data'), y=dict(field='residuals', units='data'),
                      text=dict(field='teamID', units='data'), source=source, text_color='green',
                      title='Residuals - {}'.format(year), **text_props)
    
    resid_plot.line(x='salary', y=0, source=source, line_width=2)
    
    # Change OAK color to red.
    salary = source.data['salary'][oak_pos]
    wins = source.data['W'][oak_pos]
    fit_plot.text(x=[salary], y=[wins],
                  text='OAK', text_color='red', **text_props)
    
    residual = source.data['residuals'][oak_pos]
    resid_plot.text(x=[salary], y=[residual],
                  text='OAK', text_color='red', **text_props)
    
    xaxis().axis_label='salaries (millions of Dollars)'
    yaxis().axis_label='wins above expected'
    grid().grid_line_color=None
    
    return [fit_plot, resid_plot]
```

#### Problem 2(d) 

Use exploratory data analysis tools such as histograms and boxplots to explore the distribution of the income per person by region data set from 2(c) for a given year. Describe how these change through the recent years?

**Hint**: Use a `for` loop to consider multiple years.  

```python
# Bokeh doesn't have a "boxplot" type plot, so we create one from scratch.
# The boxplot code was copied from http://bokeh.pydata.org/docs/gallery/boxplot.html and modified.
def make_boxplot(year, data):
    data.Income = data.Income/10**3
    
    cats = list(data.Region.unique())

    # Find the quartiles, IQR, and outliers for each category
    groups = data.groupby('Region')
    q1 = groups.quantile(q=0.25)
    q2 = groups.quantile(q=0.5)
    q3 = groups.quantile(q=0.75)
    iqr = q3 - q1
    upper = q2 + 1.5*iqr
    lower = q2 - 1.5*iqr

    def outliers(group):
        cat = group.name
        return group[(group.Income > upper.loc[cat][0]) | (group.Income < lower.loc[cat][0])]['Income']
    out = groups.apply(outliers).dropna()

    # Prepare outlier data for plotting, we need and x (categorical) and y (numeric)
    # coordinate for every outlier.
    outx = []
    outy = []
    for cat in cats:
        # only add outliers if they exist
        if not out.loc[cat].empty:
            for value in out[cat]:
                outx.append(cat)
                outy.append(value)

    # If no outliers, shrink lengths of stems to be no longer than the minimums or maximums
    qmin = groups.quantile(q=0.00) 
    qmax = groups.quantile(q=1.00)
    upper.Income = [min([x, y]) for (x, y) in zip(list(qmax.iloc[:, 0]), upper.Income) ]
    lower.Income = [max([x, y]) for (x, y) in zip(list(qmin.iloc[:, 0]), lower.Income) ]

    # stems
    box_plot = segment(cats, upper.Income, cats, q3.Income, x_range=cats,
                       line_width=2, line_color="black", tools='',
                       plot_width=600, plot_height=600, y_range=[0, max(groups.Income.max())*1.1],
                       title='Income ~ Region ({})'.format(year),)
    box_plot.segment(cats, lower.Income, cats, q1.Income, x_range=cats,
                     line_width=2, line_color="black")

    # boxes
    box_plot.rect(cats, (q3.Income + q2.Income)/2, 0.7, q3.Income - q2.Income,
                  fill_color="#E08E79", line_width=2, line_color="black")
    box_plot.rect(cats, (q2.Income + q1.Income)/2, 0.7, q2.Income - q1.Income,
                  fill_color="#3B8686", line_width=2, line_color="black")

    # whisters (0-height rects simpler than segments)
    box_plot.rect(cats, lower.Income, 0.2, 0, line_color="black")
    box_plot.rect(cats, upper.Income, 0.2, 0, line_color="black")

    # outliers
    box_plot.circle(outx, outy, size=6, color="#F38630", fill_alpha=0.6)

    yaxis().axis_label='income (thousands of dollars)'
    ygrid().grid_line_width = 2
    xaxis().major_label_orientation = np.pi/3
    xaxis().major_label_text_font_size="12pt"
    grid().grid_line_color=None
    
    group_incomes = groups.mean().Income
    group_regions = group_incomes.index
    hist_plot = rect(x=list(group_regions), y=group_incomes/2, width=0.8, height=group_incomes,
                     title='Average income ~ Region ({})'.format(year), tools='', plot_width=600, plot_height=600,
                     x_range=list(group_regions), y_range=[0, max(groups.Income.max())*1.1])

    yaxis().axis_label='average income (thousands of dollars)'
    ygrid().grid_line_width = 2
    xaxis().major_label_orientation = np.pi/3
    xaxis().major_label_text_font_size="12pt"
    grid().grid_line_color=None
    
    return [box_plot, hist_plot]
```

```python
reset_output()
output_file('income_distribution_over_years.html')

# gridplot expects an list of lists of plots, so create one for each year in this century.
plots = [make_boxplot(year, data=mergeByYear(year)) for year in xrange(2000, max(np.int64(income.index)))]
# Creata a grid of plotx (num. of years by 2)
gridplot(plots)

show()
```

#### Problem 3(a)

Assume you have two list of numbers, X and Y, with distribution approximately normal. X and Y have standard deviation equal to 1, but the average of X is different from the average of Y.  If the difference in the average of X and the average of Y is larger than 0, how does the proportion of X > a compare to the proportion of Y > a? 

Write a function that analytically calculates the ratio of these two proportions: Pr(X > a)/Pr(Y > a) as function of the difference in the average of X and the average of Y. 

**Hint**: Use the `scipy.stats` module for useful functions related to a normal random variable such as the probability density function, cumulative distribution function and survival function. 

**Update**: Assume Y is normally distributed with mean equal to 0. 

Show the curve for different values of a (a = 2,3,4 and 5).

```python
"""
Function
--------
ratioNormals

Return ratio of these two proportions: 
    Pr(X > a)/Pr(Y > a) as function of 
    the difference in the average of X 
    and the average of Y. 

Parameters
----------
diff : difference in the average of X 
    and the average of Y. 
a : cutoff value

Returns
-------
Returns ratio of these two proportions: 
    Pr(X > a)/Pr(Y > a)
    
Example
-------
>>> ratioNormals(diff = 1, a = 2)
"""
def ratioNormals(diff, a):
    from scipy.stats import norm
    # loc is the mean and scale the standard deviation.
    return (1-norm.cdf(a, loc=diff, scale=1))/(1-norm.cdf(a, loc=0, scale=1))

ratioNormals(diff=1, a=2)
```

Show the curve for different values of a (a = 2,3,4 and 5).

```python
reset_output()
output_file('ratios.html')

x = np.linspace(1, 2, 100)
colormap = [
    "#444444", "#a6cee3", "#1f78b4", "#b2df8a", "#33a02c", "#fb9a99",
    "#e31a1c", "#fdbf6f", "#ff7f00", "#cab2d6", "#6a3d9a"
]

hold()
for a in xrange(2, 6):
    y = [ratioNormals(diff, a) for diff in x]
    line(x, y, line_width=2, alpha=0.7, title='Ratios ~ difference',
         tools='', legend='a={}'.format(a), color=colormap[a])
    
xaxis().axis_label = 'difference'
yaxis().axis_label = 'ratio'
grid().grid_line_color=None

show()
```
