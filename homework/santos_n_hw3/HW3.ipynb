#### Problem 1(c)

Now, consider only team/season combinations in which the teams played 162 Games. Exclude all data from before 1947. Compute the per plate appearance rates for singles, doubles, triples, HR, and BB. Create a new pandas DataFrame called `stats` that has the teamID, yearID, wins and these rates.

**Hint**: Singles are hits that are not doubles, triples, nor HR. Plate appearances are base on balls plus at bats.

```python
# We need the teams for the wins. The rest can be taken from players.
# Get seasons with 162 games and starting from 1947.
#good_seasons = teams[teams["yearID"] >= 1947].groupby("yearID").sum()[["G"] == 162].index
good_seasons = (teams[teams["yearID"] >= 1947].groupby("yearID").max()["G"] == 162).index
# Filter out "bad" seasons.
seasons = teams[teams["yearID"].isin(good_seasons)][["teamID", "yearID", "W"]].reset_index(drop=True)
seasons.head()
```

```python
# Merge seasons and players before calculating singles to save cycles.
stats = seasons.merge(players[["teamID", "yearID", "H", "2B", "3B", "HR", "AB", "BB", "HBP", "SH", "SF"]]
                      .groupby(["teamID", "yearID"], as_index=False).sum(), on=["teamID", "yearID"])
# Calculate singles and rename column.
stats["H"] = stats.apply(lambda x: x["H"] - x["2B"] - x["3B"] - x["HR"], axis=1)
stats.rename(columns={"H": "1B"}, inplace=True)
# Compute rates.
# PA = AB + BB + HBP + SH + SF + Times Reached on Defensive Interference (Wikipedia).
# Note: Could not find stat for "Times Reached on Defensive Interference.
pa = stats.apply(lambda x: x["AB"] + x["BB"] + x["HBP"] + x["SH"] + x["SF"], axis=1)
rates = stats[["1B", "2B", "3B", "HR", "BB"]].divide(pa, axis=0)
# Replace old columns with new ones.
stats = stats.drop(["1B", "2B", "3B", "HR", "AB", "BB", "HBP", "SH", "SF"],
                   axis=1).merge(rates, left_index=True, right_index=True)
stats.head()
```

#### Problem 1(d)

```python
group = stats.drop(["teamID", "W"], axis=1).groupby("yearID")
group.mean().plot(figsize=(12, 8), title="Average statistics by year")
```

One noticeable trend is that the average of the statistics remained relatively constand over the years. We can also see that triples seem to have a lower variance than the rest of the statistics.

#### Problem 1(e) 

Using the `stats` DataFrame from Problem 1(c), adjust the singles per PA rates so that the average across teams for each year is 0. Do the same for the doubles, triples, HR, and BB rates. 

```python
# Standardize year group and rename columns (not really since we're not dividing by the SD, but you get the idea).
stats_std = group.transform(lambda x: x - x.mean())
stats_std.columns = ["STD_1B", "STD_2B", "STD_3B", "STD_HR", "STD_BB"]
# Merge to stats. The merge must be done on index (row number) because we dropped teamID for the group. This should,
# not be a problem since none of the operations above reorderd or changed the frame shape.
stats = stats.merge(stats_std, left_index=True, right_index=True)
stats.head()
```

#### Problem 1(f)

Build a simple linear regression model to predict the number of wins from the average adjusted singles, double, triples, HR, and BB rates. To decide which of these terms to include fit the model to data from 2002 and compute the average squared residuals from predictions to years past 2012. Use the fitted model to define a new sabermetric summary: offensive predicted wins (OPW). Hint: the new summary should be a linear combination of one to five of the five rates.


```python
g = sns.PairGrid(stats, y_vars=["W"], x_vars=["STD_1B", "STD_2B", "STD_3B", "STD_HR", "STD_BB"], size=4)
g.map(sns.regplot)
sns.despine(left=True)
```

```python
# Use only statistics after 2002 to fit.
stats_std_post_2002 = stats[stats["yearID"] >= 2002]
sm.ols("W ~ STD_1B + STD_2B + STD_3B + STD_HR + STD_BB", stats_std_post_2002).fit().summary()
```

```python
opw = sm.ols("W ~ STD_1B + STD_2B + STD_HR + STD_BB", stats_std_post_2002).fit()
opw.summary()
```

The summary of the full model shows that the triple base statistic is not significant (t-statistic > 5%), so we removed it. The new model has an adjusted $R^2$ that is equal to that of the full model, so removing the variables was probably a good idea.

The new model would be: $\text{OPW} = 80.97 + 476.44\times \text{STD_1B} + 369.41\times \text{STD_2B} + 988.53\times \text{STD_HR} + 536.43\times \text{STD_BB}$.

#### Problem 1(g)

Now we will create a similar database for individual players. Consider only player/year combinations in which the player had at least 500 plate appearances. Consider only the years we considered for the calculations above (after 1947 and seasons with 162 games). For each player/year compute singles, doubles, triples, HR, BB per plate appearance rates.
```python
# Merge seasons and players before calculating singles to save cycles.
playerstats = seasons.drop(["teamID", "W"],
                           axis=1).merge(sub_players[sub_players["PA"] >= 500], on=["yearID"])

# Calculate singles and rename column.
playerstats["H"] = playerstats.apply(lambda x: x["H"] - x["2B"] - x["3B"] - x["HR"], axis=1)
playerstats.rename(columns={"H": "1B"}, inplace=True)
# Compute rates.
rates = playerstats[["1B", "2B", "3B", "HR", "BB"]].divide(playerstats["PA"][0])
# Replace old columns with new ones.
playerstats = playerstats.drop(["1B", "2B", "3B", "HR", "BB", "PA"], axis=1).merge(rates, left_index=True, right_index=True)
playerstats.head()
```

```python
player_group = playerstats.drop("playerID", axis=1).groupby("yearID")
# Standardize year group and rename columns (not really since we're not dividing by the SD, but you get the idea).
player_std = player_group.transform(lambda x: x - x.mean())
player_std.columns = ["STD_1B", "STD_2B", "STD_3B", "STD_HR", "STD_BB"]
# Merge to stats. The merge must be done on index (row number) because we dropped teamID for the group. This should,
# not be a problem since none of the operations above reorderd or changed the frame shape.
playerstats = playerstats.merge(player_std, left_index=True, right_index=True)
playerstats.head()
```

# Problem 2:  $k$-Nearest Neighbors and Cross Validation 

What is the optimal $k$ for predicting species using $k$-nearest neighbor classification 
on the four features provided by the iris dataset.

In this problem you will get to know the famous iris data set, and use cross validation to select the optimal $k$ for a $k$-nearest neighbor classification. This problem set makes heavy use of the [sklearn](http://scikit-learn.org/stable/) library. In addition to Pandas, it is one of the most useful libraries for data scientists! After completing this homework assignment you will know all the basics to get started with your own machine learning projects in sklearn. 

Future lectures will give further background information on different classifiers and their specific strengths and weaknesses, but when you have the basics for sklearn down, changing the classifier will boil down to exchanging one to two lines of code.

The data set is so popular, that sklearn provides an extra function to load it:

```python
#load the iris data set
iris = sklearn.datasets.load_iris()

X = iris.data  
Y = iris.target

print X.shape, Y.shape
```

#### Problem 2(a) 
Split the data into a train and a test set. Use a random selection of 33% of the samples as test data. Sklearn provides the [`train_test_split`](http://scikit-learn.org/stable/modules/generated/sklearn.cross_validation.train_test_split.html) function for this purpose. Print the dimensions of all the train and test data sets you have created. 

```python
X_train, X_test, y_train, y_test = sklearn.cross_validation.train_test_split(X, Y, test_size=.33, random_state=69)
print X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

#### Problem 2(b)

Examine the data further by looking at the projections to the first two principal components of the data. Use the [`TruncatedSVD`](http://scikit-learn.org/stable/modules/generated/sklearn.decomposition.TruncatedSVD.html) function for this purpose, and create a scatter plot. Use the colors on the scatter plot to represent the different classes in the target data. 

```python
svd = sklearn.decomposition.TruncatedSVD(n_components=2, random_state=69).fit(X_train)
svd.explained_variance_ratio_.sum()
```

```python
X_reduced = svd.transform(X_train)
sns.lmplot("PC1", "PC2", data=pd.DataFrame({"PC1": X_reduced[:, 0],
                                            "PC2": X_reduced[:, 1],
                                            "Labels": y_train}),
           hue="Labels", size=8, fit_reg=False)
```

#### Problem 2(c) 

In the lecture we discussed how to use cross validation to estimate the optimal value for $k$ (the number of nearest neighbors to base the classification on). Use ***ten fold cross validation*** to estimate the optimal value for $k$ for the iris data set. 

**Note**: For your convenience sklearn does not only include the [KNN classifier](http://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html), but also a [grid search function](http://scikit-learn.org/stable/modules/generated/sklearn.grid_search.GridSearchCV.html#sklearn.grid_search.GridSearchCV). The function is called grid search, because if you have to optimize more than one parameter, it is common practice to define a range of possible values for each parameter. An exhaustive search then runs over the complete grid defined by all the possible parameter combinations. This can get very computation heavy, but luckily our KNN classifier only requires tuning of a single parameter for this problem set. 

```python
knn = sklearn.neighbors.KNeighborsClassifier()
grid = sklearn.grid_search.GridSearchCV(knn, {"n_neighbors": range(1, 21)}, cv=10)
grid.fit(X_reduced, y_train)
grid.best_params_
```

#### Problem 2(d)

Visualize the result by plotting the score results verus values for $k$. 

```python
# Point plot is better than boxplot here because the scores are very close to each other.
results = np.array([(s.parameters["n_neighbors"], s.mean_validation_score) for s in grid.grid_scores_])
sns.pointplot(results[:, 0].astype(int), results[:, 1])
```

Verify that the grid search has indeed chosen the right parameter value for $k$.

```python
# Get the lowest k with the maximum score.
i_max = results[:, 1].argmax()
best_k = results[i_max,0].astype(int)
print "Max score with k =", best_k
```

#### Problem 2(e)

Test the performance of our tuned KNN classifier on the test set.

```python
knn.n_neighbors = best_k
knn.fit(X_train, y_train)
```

```python
knn.score(X_test, y_test)
```

## Discussion for Problem 2

The first point to note is that this dataset can be very well represented by only two components. Specificaly, PC1 and PC2 account for almost 98% of the variance in the data. This helps make our calculations simpler and faster.

As for the classifier, it seems to be a great model for this dataset as shown by the 98% success rate. It should be noted that even though we used the reduced dataset to find the optimal parameters, we did not use it to train or test the mode.

```python
from sklearn.pipeline import Pipeline

estimators = [('svd', sklearn.decomposition.TruncatedSVD()), ('knn', sklearn.neighbors.KNeighborsClassifier())]
clf = Pipeline(estimators)
params = dict(svd__n_components=range(1, 11), knn__n_neighbors=range(1, 21))
grid = sklearn.grid_search.GridSearchCV(clf, param_grid=params, cv=10)
grid.fit(X_train, y_train)
grid.best_params_
```

```python
# Get score for each dimension of each trial of the optimal k.
sns.boxplot([s.cv_validation_scores for s in grid.grid_scores_
             if s.parameters["knn__n_neighbors"] == grid.best_params_["knn__n_neighbors"]])
```
